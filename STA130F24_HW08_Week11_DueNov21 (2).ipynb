A Classification Decision Tree is a predictive model used to solve classification problems, where the goal is to predict which category or class an observation belongs to. It works by repeatedly splitting the data into subsets based on specific features, creating a tree-like structure. At each split, the tree tries to maximize the separation of classes. Some real world applications are medical diagnosis: predicting whether a patient has a specific disease based on symptoms and test results; customer segmentation: identifying whether a customer is likely to buy a product or churn based on their demographic and behavioral data. 

The difference between classification deceision tree and multiple linear regression predictions lie in their prediction method. A classification deecision tree makes predictions based on their tree structure, where it starts at the root node and navigates through the tree based on feature values. At each node, it evaluates a specific feature's value and branches accordingly. At the leaf node, it assigns the class that has the majority of the data ponts in that subset. MLR makes predictions through mathematical equation, where regression uses a continuous formulas involving predicted value, input features and error term. It predicts a continuous numerical value. It calculates the coefficients to minimie the difference between the predicted values and the actual values.

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

Explaination of the Two Steps:
1. y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']:
- This line encodes the categorical variable Hard_or_Paper into a binary variable y.
- It uses pd.get_dummies to create dummy variables for each category of Hard_or_Paper and then selects the column corresponding to 'H' (hardcover).
- As a result, y is a binary target variable where:
    - 1 indicates a hardcover book.
    - 0 indicates a paperback book.
2. X = ab_reduced_noNaN[['List Price']]:
- This extracts the List Price column from the dataset as the predictor variable X.
- It ensures that X is in the format required by scikit-learn (a DataFrame with one or more columns).


Code to Train the Decision Tree

````python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Step 1: Prepare the data
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']  # Binary target variable
X = ab_reduced_noNaN[['List Price']]  # Predictor variable

# Step 2: Initialize and train the Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# Step 3: Visualize the Decision Tree
plt.figure(figsize=(10, 6))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.show()
````

Explanation of the Fitted Decision Tree
1. Tree Splits:
- The tree will divide the data based on thresholds for the List Price variable.
- At each split, the decision boundary aims to maximize the separation of hardcovers (1) and paperbacks (0).
2. Max Depth of 2:
- Limits the tree to two levels, meaning there will be at most three decision boundaries.
3. Output Visualization:
- Each node in the tree shows:
- Condition: The threshold for List Price.
- Samples: The number of observations at that node.
- Class Distribution: Proportion of hardcovers vs. paperbacks.
- Prediction: The majority class at the node.


What Predictions Are Made Based on List Price?
- If the List Price is below the first split threshold, the tree assigns books to either hardcover or paperback depending on the distribution in that branch.
- The process continues at the next level until the max_depth of 2 is reached.
- Example:
    - List Price  20: Predict paperback.
    - List Price > 20 and  50: Predict hardcover.
    - List Price > 50: Predict hardcover (if the majority in that range are hardcovers).

General Explanation of Predictions for clf2
1. How Predictions Are Made:

    - At each node, the value of the corresponding feature is compared to the threshold.
    - Depending on whether the value is below or above the threshold, the path continues to the left or right child node.
    - When a leaf node is reached, the class with the majority of samples at that node is predicted.

2. Impact of max_depth=4:
- Limits the tree to a depth of 4, meaning it can have up to 5 levels (root + 4 layers of splits).
- Ensures that the tree captures some complexity in the relationships between predictors and the target variable without overfitting.

3. Interpretation of Leaf Nodes:

````python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score

# Predictions for clf and clf2
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']
X_test_clf = ab_reduced_noNaN_test[['List Price']]
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]

y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrices
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Calculate metrics
def calculate_metrics(cm):
    TN, FP, FN, TP = cm.ravel()
    sensitivity = TP / (TP + FN)  # True Positive Rate
    specificity = TN / (TN + FP)  # True Negative Rate
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    return sensitivity, specificity, accuracy

sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(cm_clf)
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(cm_clf2)

# Report results
print(f"Confusion Matrix for clf:\n{cm_clf}")
print(f"Sensitivity: {sensitivity_clf:.2f}, Specificity: {specificity_clf:.2f}, Accuracy: {accuracy_clf:.2f}")

print(f"\nConfusion Matrix for clf2:\n{cm_clf2}")
print(f"Sensitivity: {sensitivity_clf2:.2f}, Specificity: {specificity_clf2:.2f}, Accuracy: {accuracy_clf2:.2f}")
````

To visualize feature importances for scikit-learn classification decision trees and interpret the results for `clf2`, follow these steps:

1. Use the `.feature_importances_` attribute of the trained decision tree model `clf2` to access the importance scores of each feature.
2. The `.feature_names_in_` attribute provides the names of the features corresponding to these importance scores.
3. Pair the scores with their respective feature names and sort them to determine which feature contributes the most to the predictions.

Heres how you can do it programmatically and interpret the results:


````python
import matplotlib.pyplot as plt
import numpy as np

# Access feature importances and feature names
importances = clf2.feature_importances_
feature_names = clf2.feature_names_in_

# Pair and sort them
sorted_indices = np.argsort(importances)[::-1]  # Descending order
sorted_importances = importances[sorted_indices]
sorted_feature_names = feature_names[sorted_indices]

# Visualize the feature importances
plt.barh(sorted_feature_names, sorted_importances)
plt.xlabel("Feature Importance")
plt.title("Feature Importances for clf2")
plt.gca().invert_yaxis()  # For descending order
plt.show()

# Print the most important feature
print(f"The most important predictor variable for clf2 is '{sorted_feature_names[0]}'.")
````

This visualization will provide an intuitive understanding of how much each feature contributes to the predictive power of `clf2`. The bar plot ranks features by importance, and the output specifies which feature is most significant.
